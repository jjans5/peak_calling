# Run code in parallel on all species

```bash
# List of species and matching chain file names
declare -A chains
chains["Bonobo"]="panPan2ToHg38"
chains["Chimpanzee"]="panTro5ToHg38"
chains["Gorilla"]="gorGor4ToHg38"
chains["Macaque"]="rheMac10ToHg38"

for species in "${!chains[@]}"; do
  input_file="fragment_files/${species}_enterocytes.fragments.tsv.gz"
  chain_file="/cluster/project/treutlein/USERS/jjans/analysis/adult_intestine/atac/genomes/chains/${chains[$species]}.over.chain.gz"
  output_file="lifted_fragments/${species}_enterocytes.hg38.fragments_par.tsv.gz"

  echo "üöÄ Running liftover for $species..."
  bash liftover_fragments_par.sh --i "$input_file" --c "$chain_file" --o "$output_file" --ncpu 30 &> "lifted_fragments/${species}.log"
  echo "‚úÖ Done with $species"
  echo
done
```

```bash
# === Loop through species ===
for species in "Marmoset"; do
  input_frag="lifted_fragments/${species}_enterocytes.hg38.fragments_par.tsv.gz"
  chrom_file="/cluster/home/jjanssens/jjans/analysis/cerebellum/genomes_new/homo_sapiens/hg38.chrom.sizes"
  output_bw="bigwigs/${species}_enterocytes.hg38.bw"

  if [[ -s "$input_frag" && -s "$chrom_file" ]]; then
    echo "üéØ Converting $species to bigWig..."
    scatac_fragment_tools bigwig \
      -i "$input_frag" \
      -c "$chrom_file" \
      -o "$output_bw" \
      -x 
    echo "‚úÖ Done: $output_bw"
  else
    echo "‚ö†Ô∏è  Missing input or chrom sizes for $species ‚Äî skipping"
  fi
done
```
